```python
# Importing Necessary Libraries
import pandas as pd
import yfinance as yf
from datetime import datetime
from IPython.display import display
```

```python
# Downloading data for the MAANG Companies
Company_names = ['Amazon', 'Apple', 'Google', 'Meta', 'Netflix']
company_dict = {}
company_dict["Amazon"]=yf.download('AMZN', period='max')
company_dict["Apple"]=yf.download('AAPL', period='max')
company_dict["Google"]=yf.download('GOOG', period='max')
company_dict["Meta"]=yf.download('META', period='max')
company_dict["Netflix"]=yf.download('NFLX', period='max')
```

```python
# Display Dataframes
for i in Company_names:
  print()
  print(f"Dataframe for {i} Company")
  display(company_dict[i])
  print()
```

```python
# Describe Dataframes with null values
for i in Company_names:
  print()
  print(f"Description of data for {i} Company")
  display(company_dict[i].describe())
  print(f"Null values in the data for {i} Company")
  display(company_dict[i].isnull().sum(axis = 0))
  print()
```

```python
# Saving the Scarpped Data
for i in Company_names:
  company_dict[i].to_csv(i+'.csv', index=True)
```
