```python
# Importing Necessary Libraries
import pandas as pd
```

```python
# Function to change Date column to datetime object
def obj_to_datetime(df):
  df['Date'] = pd.to_datetime(df['Date'])
  return df
```

```python
# Loading the scrapped data
Company_names = ['Amazon', 'Apple', 'Google', 'Meta', 'Netflix']
company_dict = {}
for i in Company_names:
  company_dict[i]=obj_to_datetime(pd.read_csv(i+'.csv')).set_index('Date')
```

```python
# Importing Necessary Libraries
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
# Setting up the Batch Size and Sequence Length for future stock price prediction
batch_size = 64
f_len = 10
```

```python
# Function to get a transformed data with MinMax Scalling and labels
def construct_feature_target(df_in, f_len=3, batch_size = 64):
  df_column_list = []
  for i in range(f_len):
    df_column_list.append("Feature"+str(i+1))
  # print(df_column_list)
  df = df_in['Close'].to_numpy().reshape(-1, 1)
  sc = MinMaxScaler(feature_range=(0,1))
  df = sc.fit_transform(df)
  nbatch = (len(df)-f_len)//batch_size
  lendf = nbatch*batch_size + f_len
  df = df[-lendf:]
  X = []
  y = []
  df_in = df_in[-lendf:]
  df_in = df_in[f_len:]
  for i in range(f_len,len(df)):
    X.append(df[(i-f_len):i])
    y.append(df[i])
  X = np.array(X)
  y = np.array(y)
  X = np.reshape(X,(X.shape[0],X.shape[1]))
  y = np.reshape(y,(y.shape[0],y.shape[1]))
  df = pd.DataFrame(data=X, index=None, columns=df_column_list)
  df["Date"] = df_in.index
  df = df.set_index("Date")
  df["Target"] = y
  return df
```

```python
# Saving the transformed Data
for i in Company_names:
  df = construct_feature_target(company_dict[i], f_len=f_len, batch_size=batch_size)
  df.to_csv(i+'_df.csv', index=True)
```
